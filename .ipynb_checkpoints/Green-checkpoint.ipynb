```python

from pystac_client import Client
import planetary_computer as pc

# Search against the Planetary Computer STAC API
catalog = Client.open(
  "https://planetarycomputer.microsoft.com/api/stac/v1"
)

# Define your area of interest
aoi = {
  "type": "Polygon",
  "coordinates": [
    [
      [4.7040202453788424, 52.26657736374196],
      [5.212675109079015, 52.26657736374196],
      [5.212675109079015, 52.44441426462788],
      [4.7040202453788424, 52.44441426462788],
      [4.7040202453788424, 52.26657736374196]
    ]
  ]
}

daterange = {"interval": ["2022-06-01T00:00:00Z", "2022-08-30T23:59:59Z"]}
# Define your search with CQL2 syntax
search = catalog.search(filter_lang="cql2-json", filter={
  "op": "and",
  "args": [
    {"op": "s_intersects", "args": [{"property": "geometry"}, aoi]},
    {"op": "anyinteracts", "args": [{"property": "datetime"}, daterange]},
    {"op": "=", "args": [{"property": "collection"}, "landsat-c2-l2"]},
    {"op": "<=", "args": [{"property": "eo:cloud_cover"}, 10]}
  ]
})

# Grab the first item from the search results and sign the assets
first_item = next(search.get_items())
pc.sign_item(first_item).assets
```

```python
items = search.get_all_items()
```

```python
ds = stackstac.stack(pc.sign(items), epsg=28992)
ds
```

```python
xmin, xmax, ymin, ymax = 110189.0 , 476708.7 , 135933.0 , 493892.9  # Set to a small area to limit computation time.
amsterdam = ds.loc[:,:, ymax:ymin,xmin:xmax]
amsterdam
```

```python
red = amsterdam.sel(band='red')
nir = amsterdam.sel(band='nir08')
ndvi = (nir-red) / (nir+red)
monthly = ndvi.resample(time="MS").max("time", keep_attrs=True)
monthly
```

```python
import rich.table
import dask.diagnostics

with dask.diagnostics.ProgressBar():
    ts = monthly.compute()
```
